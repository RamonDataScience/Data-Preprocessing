```python
# Calculamos el promedio y desv. estan. de las columnas aún cuando se tienen valores perdidos
```

```python
dfcontaminacion.mean()# Calculamos el promedio de las variables del objeto "dfcontaminacion"
```

```python
dfcontaminacion.std() # Calculamos desv. estand. de las variables del objeto "dfcontaminacion"
```

# REEMPLAZANDO LOS VALORES PERDIDOS

```python
# REEMPLAZO DE LOS VALORES PERDIDOS CON UN VALOR FIJO

dfcontaminacion1 = dfcontaminacion
```

```python
dfcontaminacion1.fillna(0) # Reemplazamos los valores faltantes con un zero
```

```python
# Reemplazo de los valores perdidos con "la media (promedio)" de cada columna

dfcontaminacion2 = dfcontaminacion.apply(lambda x: x.fillna(x.mean()), axis=0)
# axis = 0 para indicar que lo que me interesa es trabajar con las columnas
# x.mean() es para obtener el valor promedio de cada columna: x representa cada una de las columnas que va a ir recorriendo la funcion apply
# lambda x: es una funcion anónima(no tiene un nombre definido) sirve para realizar funciones simples, los calculos que realiza aparecen después de los dos puntos (el llenado de los valores perdidos de cada columna con los valores promedio de la misma) 
dfcontaminacion2
```

```python
# REEMPLAZAR VALORES PERDIDOS CON LA "MEDIANA" (POSICION CENTRAL CUANDO SE ORDENA UN CONJUNTO DE DATOS) DE CADA COLUMNA
```

```python
dfcontaminacion3 = dfcontaminacion.apply(lambda x: x.fillna(x.median()), axis=0)
 
dfcontaminacion3
```

# EJEMPLO CON VARIABLES CATEGÓRICAS

```python
dfcompras = pd.read_csv('Patron de Compra con NAs.csv')
```

```python
dfcompras
```

```python
dfcompras = pd.read_csv('Patron de Compra con NAs.csv', index_col=0)
dfcompras
```

```python
dfcompras.info() # Python almacena los datos categóricos como objetos.
```

```python
dfcompras
```

```python
# Reemplazamos los valores perdidos de la columna Profesión con la palabra "Desconocido"
```

```python
dfcompras['Profesion'].fillna("Desconocida", inplace = True)
dfcompras
```

```python
# Reemplazando con los valores de la fila anterior
```

```python
dfcompras2 = dfcompras.fillna(method = 'ffill')
dfcompras2
```

```python
# Reemplazo de valores faltantes con la siguiente fila
```

```python
dfcompras3 = dfcompras.fillna(method = 'bfill')
dfcompras3
```
