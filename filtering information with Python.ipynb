```python
import pandas as pd
```

```python
# Leemos el archivo a analizar
```

```python
atletas = pd.read_csv('Corredores.csv')
atletas.info() # Vemos el tipos de datos que tenemos en el Dataframe
```

```python
atletas.head()
```

```python
# Obtenemos las primeras quince filas
```

```python
atletas.iloc[0:15]
```

```python
# Si queremos que nos traiga las primeras tres columnas
```

```python
atletas.iloc[ : , 0:3]
```

```python
# Para traer columnas que NO son consecutivas
```

```python
atletas.iloc[ :, [0,3,4]]
```

```python
# Si queremos solamente determinadas filas y columnas
```

```python
atletas.iloc[[0,5,7], [0,3,4]]
```

```python
# Obtenemos solamente el Género y Tiempo de los Mexicanos y Argentinos
```

```python
atletas.loc[["MEX","ARG"],["Genero","Tiempo"]]
```

```python
# Desplegamos un rango de columnas
```

```python
atletas.loc[50, "Genero":"Tiempo"]
```

```python
# Obteniendo a las mujeres que participaron en la maratón
```

```python
atletas_Mujeres = atletas[atletas["Genero"]=="Female"]
```

```python
atletas_Mujeres
```

```python
atletas_Mujeres.info()
```

```python
# Obtenemos los atletas mayores de 40 años
```

```python
atletas_Edad = atletas[atletas["Edad"] > 40]
atletas_Edad
```

```python
# Buscando participantes mujeres o participantes cuyo tiempo fué menor a 200 minutos
```

```python
atletas_Mujeres_o_Veloz = atletas[(atletas["Genero"]== 'Female') | (atletas["Tiempo"] < 200)]
atletas_Mujeres_o_Veloz
```

```python
# Obtenemos la información de las mujeres mayores de 40 años
```

```python
atletas_Mujeres_Edad = atletas[(atletas["Genero"] == "Female") & (atletas["Edad"] > 40)]
atletas_Mujeres_Edad
```

```python
# Guardando los datos en un archivo csv
```

```python
atletas_Mujeres_Edad.to_csv("atletas_Mujeres_Edad.csv")
```
