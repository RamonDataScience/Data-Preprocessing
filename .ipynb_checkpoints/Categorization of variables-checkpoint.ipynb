# Categorización de variables

```python
# Importamos librerias
```

```python
import pandas as pd
import math
```

```python
atletas = pd.read_csv('Corredores.csv')
```

```python
atletas.info()
```

```python
atletas.head()
```

```python
atletas.describe()
```

```python
# Definimos los limites que tendra cada categoria de la variable Tiempo
```

```python
valores = [-math.inf, 200, 300, math.inf]
```

```python
# Indicamos cuales seran los nombres de las 3 categorías (<200, 200-300, >300)
```

```python
categorias = ["Veloz", "Regular", "Lento"]
```

```python
# Agregamos una columna nueva en la que estarán indicadas las categorias asignadas
```

```python
atletas['Velocidad'] = pd.cut(x=atletas['Tiempo'], bins = valores, labels = categorias)
atletas
```
